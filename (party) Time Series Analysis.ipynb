# BJP party tweet


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("BJP1.csv")

# Convert the 'created_at' column to datetime format
df['created_at'] = pd.to_datetime(df['Date'])

# Group the data by member and time period (in this case, by day)
grouped = df.groupby(['User', pd.Grouper(key='created_at', freq='D')]).size().reset_index(name='count')

# Pivot the data to have the member names as columns
pivot = grouped.pivot(index='created_at', columns='User', values='count')

# Resample the data to fill in any missing dates and aggregate by sum
resampled = pivot.resample('D').sum()

# Plot the time series for each member
resampled.plot(figsize=(10,6))
plt.xlabel('Date')
plt.ylabel('Tweet Frequency')
plt.title('Tweet Frequency of Party Members Over Time')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('BJP1.csv')

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['Date'])

# Group data by member and date, and count the number of tweets
member_data = data.groupby(['User', pd.Grouper(key='date', freq='M')]).size().reset_index(name='count')

# Loop over each member and plot their tweet frequency over time
fig, ax = plt.subplots(figsize=(10,6))
for member in member_data['User'].unique():
    member_df = member_data[member_data['User']==member]
    ax.plot(member_df['date'], member_df['count'], label=member)
    
ax.set_xlabel('Date')
ax.set_ylabel('Number of tweets')
ax.legend()
plt.show()
```
