```python
import time
import logging
import pandas as pd
from googletrans import Translator
from retrying import retry

# Initialize the translator
translator = Translator()

# Configure logging
logging.basicConfig(level=logging.INFO)

@retry(wait_fixed=5000, stop_max_attempt_number=3)
def translate_batch(batch):
    try:
        # Translate the Hindi text in the batch to English
        batch['newTweet'] = batch['Tweet'].apply(lambda x: translator.translate(x, dest='en').text)
    except Exception as e:
        logging.error(f"Error while translating batch: {e}")
        raise

# Define the translation function
def translate(tweets_df, batch_size=100):
    # Split the DataFrame into batches of size batch_size
    batches = [tweets_df[i:i+batch_size] for i in range(0, len(tweets_df), batch_size)]

    # Translate each batch separately
    for i, batch in enumerate(batches):
        logging.info(f"Translating batch {i+1} of {len(batches)}")
        translate_batch(batch)
        # Introduce a delay of 10 seconds between batches to avoid hitting the rate limits
        time.sleep(10)
    
    # Combine the translated batches into a single DataFrame and return it
    return pd.concat(batches)






```

```python
#### below code had The code tries several values of k (the number of topics to generate) and computes the coherence
####score for each value of k. Coherence score measures the quality of the topics generated by the model. The model
####that generates topics with the highest coherence score is chosen as the optimal model.
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import NMF
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# # Load the preprocessed CSV file into a pandas dataframe
df = pd.read_csv('congressTrans.csv')

# # Remove rows with missing values in the preprocessed text column
df = df.dropna(subset=['tokenize_text'])

# # Get the preprocessed text as a list
preprocessed_text = df['tokenize_text'].tolist()

 # Create a document-term matrix using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
doc_term_matrix = vectorizer.fit_transform(preprocessed_text)

# Normalize the document-term matrix
doc_term_matrix_normalized = normalize(doc_term_matrix)

# Set the number of topics to try
num_topics = [9]   # out of many tries from 1 to 13 ,9 has the best coherencescore

# Initialize lists to store coherence scores and models
coherence_scores = []
models = []

# Iterate over the number of topics
for k in num_topics:
    # Create an NMF model with k topics
    nmf_model = NMF(n_components=k, random_state=42)
    nmf_model.fit(doc_term_matrix_normalized)

    # Compute the coherence score
    similarity_matrix = cosine_similarity(nmf_model.components_)
    coherence_score = np.mean([np.amax(similarity_matrix[i][np.nonzero(similarity_matrix[i])]) for i in range(k)])
    coherence_scores.append(coherence_score)
    
    # Store the model
    models.append(nmf_model)

# Find the index of the maximum coherence score
max_coherence_score_index = np.argmax(coherence_scores)

# Print the optimal number of topics and the corresponding coherence score
print('Optimal Number of Topics:', num_topics[max_coherence_score_index])
print('Coherence Score:', coherence_scores[max_coherence_score_index])

# Print the topics
for topic_index, topic in enumerate(models[max_coherence_score_index].components_):
    top_words_indices = topic.argsort()[:-10 - 1:-1]
    top_words = [vectorizer.get_feature_names()[i] for i in top_words_indices]
    print('Topic {}: {}'.format(topic_index, top_words))
```
