```python
import csv

def find_duplicate_tweets(csv_file):
    tweets = set()
    duplicates = []

    with open(csv_file, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row

        for row in reader:
            tweet = row[2]  # Assuming the tweet content is in the second column (index 1)
            
            # Check if the tweet is already in the set
            if tweet in tweets:
                duplicates.append(tweet)
            else:
                tweets.add(tweet)

    return duplicates

# Provide the path to your CSV file
csv_file = 'processed_tweet.csv'

# Call the function to find duplicate tweets
duplicate_tweets = find_duplicate_tweets(csv_file)

# Print the duplicate tweets
if duplicate_tweets:
    print("Duplicate tweets found:")
    for tweet in duplicate_tweets:
        print(tweet)
else:
    print("No duplicate tweets found.")
```

```python
import tweepy
import csv
import time
import os
from urllib3.exceptions import ProtocolError

# Add your API credentials here
consumer_key = 'qtd1m56Qd2bNxoPkhjDp65vcs'
consumer_secret = 'cyQtvctevq663bbyyP1lglWHaeQVQSMHCo3jLgZnzGtY7qHdcS'
access_token = '1556319879216517122-xirEJ0rMwT2gaYLTqpJY2TWVNDOCSf'
access_token_secret = 'SCG14Qvwoou00QYSi4IiUnkQV12tq9uze7Zrm9uuEPAL4'

# Authenticate with the API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
def get_tweet_metrics(tweet_id):
    try:
        tweet = api.get_status(tweet_id, tweet_mode='extended')
        retweets = tweet.retweet_count
        favorites = tweet.favorite_count
        replies_count = 0
        for _ in tweepy.Cursor(api.search_tweets, q=f"to:{tweet.user.screen_name} since_id:{tweet.id} -filter:retweets").items():
            replies_count += 1
        return retweets, favorites, replies_count
    except tweepy.TweepError as e:
        if e.api_code == 179:  # Forbidden error
            print(f"Tweet {tweet_id} is forbidden or unauthorized. Skipping...")
            return None, None, None
        else:
            raise


def main():
    # Create the file if it doesn't exist
    with open('last_processed_tweet_id1.txt', 'w') as f:
        pass

    # Read the ID of the last processed tweet from the file
    if os.path.exists('last_processed_tweet_id1.txt'):
        with open('last_processed_tweet_id1.txt', 'r') as id_file:
            last_processed_id = id_file.read().strip()
    else:
        last_processed_id = None

    # Process the tweets
    with open('BJPdup.csv', 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)  # skip the header row

        processed_tweets = []

        for row in reader:
            # Extract the tweet ID from the CSV file
            tweet_id = row[0]

            # Check if we've already processed this tweet or if we need to resume processing
            if last_processed_id is not None and tweet_id != last_processed_id:
                continue
            else:
                last_processed_id = None  # reset the ID so we process all remaining tweets

            # Get the tweet object from Twitter's API
            try:
                retweets, favorites, replies_count = get_tweet_metrics(tweet_id)
            except tweepy.TweepError as e:
                if e.api_code == 179:  # Forbidden error
                    print(f"Tweet {tweet_id} is forbidden or unauthorized. Skipping...")
                    continue
                else:
                    raise

            # Append the data to a list of processed tweets
            processed_tweets.append((tweet_id, retweets, favorites, replies_count))
            processed_count += 1

            # Save the processed data after every 100 tweets
            if processed_count % 100 == 0:
                with open('tweets_with_metrics.csv', 'a', encoding='utf-8', newline='') as g:
                    writer = csv.writer(g)
                    writer.writerows(processed_tweets)
                processed_tweets = []

            # Print progress every 10 batches
            if processed_count % (batch_size * 1) == 0:
                print(f"{processed_count} tweets processed")

        # Save the processed data after every batch
        with open('tweets_with_metrics.csv', 'a', encoding='utf-8', newline='') as g:
            writer = csv.writer(g)
            writer.writerows(processed_tweets)
        processed_tweets = []


if __name__ == '__main__':
    main()
```
