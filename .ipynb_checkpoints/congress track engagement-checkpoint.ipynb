```python
import csv
import tweepy
import os
import time
import pandas as pd
from urllib3.exceptions import ProtocolError

# Authenticate to Twitter
consumer_key = 'qtd1m56Qd2bNxoPkhjDp65vcs'
consumer_secret = 'cyQtvctevq663bbyyP1lglWHaeQVQSMHCo3jLgZnzGtY7qHdcS'
access_token = '1556319879216517122-xirEJ0rMwT2gaYLTqpJY2TWVNDOCSf'
access_token_secret = 'SCG14Qvwoou00QYSi4IiUnkQV12tq9uze7Zrm9uuEPAL4'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

# Create API object
api = tweepy.API(auth)

# Define your search query
hashtags = ['#loksabha2014','#Elections2014','#loksabha2014','#RahulGandhi','#Congress','#BJP','#elections2014','#NarendraModi','#myvote2014']
keywords = ["2014GeneralElection", "election", "Election", "ELECTION", "BJP", "bjp", "Bjp", "CONGRESS", "Congress", "congress", "RAHUL GANDHI", "rahul gandhi", "Rahul Gandhi", "Narendra Modi", "narendra modi", "NARENDRA MODI"]
query = f"{' OR '.join(hashtags)} {' OR '.join(keywords)} lang:en"



# Define your time frame
start_date = "2013-11-1"
end_date = "2014-05-7"

# Define the number of tweets you want to collect
num_tweets = 8000

# Define the time interval for saving the data to CSV (in seconds)
save_interval = 60  # 1 minute

# Define the file names for saving the data and the last processed tweet ID
file_name = 'userEng1.csv'
id_file_name = 'my_last_tweet_id.txt'

# Load the last processed tweet ID, or start from the beginning if the file doesn't exist
last_processed_id = 0
if os.path.exists('my_last_tweet_id.txt'):
    with open('my_last_tweet_id.txt', 'r') as f:
        last_processed_id_str = f.read().strip()
        if last_processed_id_str and last_processed_id_str.isdigit():
            last_processed_id = int(last_processed_id_str)

# Create a dictionary to store the data
data = {'date': [], 'time': [], 'tweet': []}

# Collect tweets using the search API
processed_count = 0
save_count = 0
processed_tweets = []

while len(data['tweet']) < num_tweets:
    # Collect tweets using the search API
    try:
        # Fetch tweets based on the last processed tweet ID or from the beginning
        if last_processed_id == 0:
            tweets = tweepy.Cursor(api.search_tweets, q=query, lang="en", since_id=start_date, until=end_date).items(num_tweets)
        else:
            max_id = last_processed_id - 1
            tweets = tweepy.Cursor(api.search_tweets, q=query, lang="en", since_id=start_date, until=end_date, max_id=max_id).items(num_tweets)

        for tweet in tweets:
            last_processed_id = tweet.id
            data['date'].append(tweet.created_at.date())
            data['time'].append(tweet.created_at.time())
            data['tweet'].append(tweet.text)

            processed_count += 1

            # Append the processed tweet to the list
            processed_tweets.append({'date': tweet.created_at.date(), 'time': tweet.created_at.time(), 'tweet': tweet.text})

            # Save the processed data after every 100 tweets
            if processed_count % 100 == 0:
                with open(f'userEng1.csv', 'w', encoding='utf-8', newline='') as g:
                    writer = csv.DictWriter(g, fieldnames=['date', 'time', 'tweet'])
                    writer.writeheader()
                    writer.writerows(processed_tweets)
                processed_tweets = []
                save_count += 1
                print(f"Processed tweets saved to processed_tweets_{save_count}.csv")

            # Stop the loop if the desired number of tweets has been collected
            if len(data['tweet']) >= num_tweets:
                break

        # Save the last processed tweet ID to a file
        with open(id_file_name, 'w') as id_file:
            id_file.write(str(last_processed_id))

    except tweepy.TooManyRequests as e:
        print('Rate limit reached. Waiting for 15 minutes...')
        time.sleep(15 * 60)  # Wait for 15 minutes if rate limit is reached
    except tweepy.TweepError as e:
        if e.response and e.response.status_code == 401:
            print("Authorization error occurred. Please check your API credentials.")
            break  # Terminate the loop if authorization error occurs
        elif e.response and e.response.status_code == 403:
            print("Forbidden error occurred. Skipping the tweet.")
            continue  # Skip to the next tweet if forbidden error occurs
        else:
            print(f"Error fetching tweets: {e}")
            continue

        
    except Exception as e:
        print(f"Error occurred: {e}")
        continue

# Save the remaining processed tweets
if processed_tweets:
    with open(f'userEng1.csv', 'w', encoding='utf-8', newline='') as g:
        writer = csv.DictWriter(g, fieldnames=['date', 'time', 'tweet'])
        writer.writeheader()
        writer.writerows(processed_tweets)
    print(f"Processed tweets saved to processed_tweets_{save_count}.csv")

# Save the data to CSV
df = pd.DataFrame(data)
df.to_csv(file_name, index=False)
print(f'{len(data["tweet"])} tweets saved to {file_name}')

```

```python
import tweepy
import csv
import time
import os
from urllib3.exceptions import ProtocolError

# Add your API credentials here
consumer_key = 'qtd1m56Qd2bNxoPkhjDp65vcs'
consumer_secret = 'cyQtvctevq663bbyyP1lglWHaeQVQSMHCo3jLgZnzGtY7qHdcS'
access_token = '1556319879216517122-xirEJ0rMwT2gaYLTqpJY2TWVNDOCSf'
access_token_secret = 'SCG14Qvwoou00QYSi4IiUnkQV12tq9uze7Zrm9uuEPAL4'

# Authenticate with the API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

# Open the CSV file containing the tweets
with open('congressdup.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)  # skip the header row
    
    # Read the ID of the last processed tweet from the file, if it exists
    last_processed_id2 = None
    if os.path.exists('last_processed_tweet_id2.txt'):
        with open('last_processed_tweet_id2.txt', 'r') as id_file:
            last_processed_id2 = id_file.read().strip()
    else:
        with open('last_processed_tweet_id2.txt', 'w') as f:
            pass 
    
    batch_size = 500  # number of tweets to process in each batch
    batches = [row for row in reader]  # split the data into batches
    batch_count = 0
    processed_count = 0
    processed_tweets = []
    for i in range(0, len(batches), batch_size):
        batch = batches[i:i+batch_size]  # get a batch of tweets
        
        # Process the tweets in the batch
        for row in batch:
            # Extract the tweet ID from the CSV file
            tweet_id = row[0]
            
            # Check if we've already processed this tweet or if we need to resume processing
            if last_processed_id2 is not None and tweet_id != last_processed_id2:
                continue
            else:
                last_processed_id2 = None  # reset the ID so we process all remaining tweets
            
            # Get the tweet object from Twitter's API
            
            
             
            try:
                tweet = api.get_status(tweet_id, tweet_mode='extended')
                 
                #Extract the retweet count and favorite count from the tweet object      
                retweets = tweet.retweet_count
                favorites = tweet.favorite_count
            
                # Search for replies to the tweet and count the number of results
                query = f"to:{tweet.user.screen_name} since_id:{tweet.id} -filter:retweets"
                replies_count = 0
                for _ in tweepy.Cursor(api.search_tweets, q=query, tweet_mode='extended').items():
                    replies_count += 1
                
                # Append the data to a list of processed tweets
                processed_tweets.append(row + [retweets, favorites, replies_count])
                processed_count += 1
             
                # Save the processed data after every 100 tweets
                if processed_count % 100 == 0:
                    with open('tweets_with_metrics2.csv', 'a', encoding='utf-8', newline='') as g:
                        writer = csv.writer(g)
                        writer.writerows(processed_tweets)
                    processed_tweets = []
            
               # Print progress every 10 batches
                if processed_count % (batch_size * 1) == 0:
                    print(f"{processed_count} tweets processed")
               
            
                    
             
            except tweepy.errors.TweepyException as e:
                if e.response is not None:
                    if e.response.status_code == 88:
                        print("Rate limit exceeded. Waiting for 15 minutes...")
                        time.sleep(15 * 60)  # wait for 15 minutes before retrying
                        continue
                    elif e.response.status_code == 179:
                        print(f"Tweet {tweet_id} is forbidden or unauthorized. Skipping...")
                        continue
                
                with open('last_processed_tweet_id2.txt', 'w') as id_file:
                    id_file.write(tweet_id)
                print(f"Error fetching tweet {tweet_id}: {e}")
                continue
            
                
        # Save the processed data after every batch
        with open('tweets_with_metrics2.csv', 'a', encoding='utf-8', newline='') as g:
            writer = csv.writer(g)
            writer.writerows(processed_tweets)
        processed_tweets = []
        batch_count += 1
        print(f"Batch {batch_count} complete")
```
