```python
## what the differnce bwtween topic modeling nad text classification

## Answer : Text classification and topic modeling are related but different tasks in natural language processing.

#Text classification is the task of automatically assigning predefined categories or labels to a text based on its content. 
# The goal is to build a model that can accurately classify new texts into one or more predefined categories.
#Topic modeling, on the other hand, is an unsupervised learning task that involves discovering latent topics or 
#themes in a collection of texts without predefined categories or labels. The goal is to identify the underlying topics that are present in the corpus and to represent each document as a mixture of these topics.
#In summary, text classification is a supervised learning task, while topic modeling is an unsupervised learning task. Text classification requires predefined categories or labels, while topic modeling does not.
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle


# Load the labeled data file with 100 rows
labeled_data = pd.read_csv('BJPtextkAnalysis.csv',na_filter= False)

# Fill any missing values in the 'tokenize_text' column with an empty string
labeled_data.dropna(subset=['label'], inplace=True)



# Load the unlabeled data file with 9600 rows
unlabeled_data = pd.read_csv('unlabeled_data.csv')

# Fill any missing values in the 'tokenize_text' column with an empty string
unlabeled_data['tokenize_text'] = unlabeled_data['tokenize_text'].fillna('')

# Define the vectorizer to use
vectorizer = TfidfVectorizer()

# Vectorize the labeled data
labeled_vectors = vectorizer.fit_transform(labeled_data['tokenize_text'])

# Define the classifier to use
classifier = LogisticRegression()

# Train the classifier on the labeled data
classifier.fit(labeled_vectors, labeled_data['label'])

# Define the number of samples to query at each iteration
batch_size = 100

# Loop until all unlabeled samples have been labeled
while len(unlabeled_data) > 0:
    
    # Vectorize the unlabeled data
    unlabeled_vectors = vectorizer.transform(unlabeled_data['tokenize_text'])
    
    # Make predictions on the unlabeled data
    predictions = classifier.predict(unlabeled_vectors)
    
    # Compute the confidence of the predictions
    confidence = classifier.predict_proba(unlabeled_vectors).max(axis=1)
    
    # Select the samples with the highest confidence
    indices = confidence.argsort()[::-1][:batch_size]
    num_samples = len(unlabeled_data)
    if len(indices) > num_samples:
        indices = indices[:num_samples]
    samples = unlabeled_data.iloc[indices]

    # Label the selected samples
    for i in indices:
        if i >= len(samples):
            continue
        print(f"Sample {i}: {samples['text'].iloc[i]}")
        label = input("Enter label (achievements and accomplishments/Political Announcement,Engagement/ responses to opposition,people/stance on national issues): ")
        samples.at[samples.index[i], 'label'] = label
    
    # Add the labeled samples to the labeled data
    labeled_data = pd.concat([labeled_data, samples], ignore_index=True)
    
    # Remove the labeled samples from the unlabeled data
    unlabeled_data.drop(samples.index, inplace=True)
    
    # Shuffle the labeled data
    labeled_data = shuffle(labeled_data)
    
    # Vectorize the updated labeled data
    labeled_vectors = vectorizer.transform(labeled_data['tokenize_text'])
    
    # Train the classifier on the updated labeled data
    classifier.fit(labeled_vectors, labeled_data['label'])
    
    # Evaluate the classifier on the labeled data
    X_train, X_test, y_train, y_test = train_test_split(labeled_vectors, labeled_data['label'], test_size=0.2, random_state=42)
    y_pred = classifier.predict(X_test)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(f"F1 score: {f1}")


```

```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Load the dataset and extract features
newsgroups_train = fetch_20newsgroups(subset='train', categories=['alt.atheism', 'sci.space'])
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(newsgroups_train.data)

# Train a classifier
classifier = LogisticRegression()
classifier.fit(vectors, newsgroups_train.target)

# Predict the labels for the test data
newsgroups_test = fetch_20newsgroups(subset='test', categories=['alt.atheism', 'sci.space'])
unlabeled_vectors = vectorizer.transform(newsgroups_test.data)
predicted_labels = classifier.predict(unlabeled_vectors)

# Select the top k most uncertain predictions for labeling
k = 1000
top_confidence = sorted(classifier.predict_proba(unlabeled_vectors).max(axis=1), reverse=True)[:k]
top_indices = [list(classifier.predict_proba(unlabeled_vectors)).index(list(filter(lambda x: max(x) == conf, classifier.predict_proba(unlabeled_vectors)))) for conf in top_confidence]
new_labeled_vectors = unlabeled_vectors[top_indices]
new_labeled_labels = predicted_labels[top_indices]

# Add the newly labeled data to the training set
labeled_vectors = np.concatenate((vectors, new_labeled_vectors))
labeled_labels = np.concatenate((newsgroups_train.target, new_labeled_labels))

# Retrain the classifier with the new data
classifier.fit(labeled_vectors, labeled_labels)

# Evaluate the performance on the test data
test_vectors = vectorizer.transform(newsgroups_test.data)
test_labels = newsgroups_test.target
accuracy = classifier.score(test_vectors, test_labels)
print("Accuracy on the test data:", accuracy)



```

```python
import tweepy
import csv
import time
import os
from urllib3.exceptions import ProtocolError

# Add your API credentials here
consumer_key = 'qtd1m56Qd2bNxoPkhjDp65vcs'
consumer_secret = 'cyQtvctevq663bbyyP1lglWHaeQVQSMHCo3jLgZnzGtY7qHdcS'
access_token = '1556319879216517122-xirEJ0rMwT2gaYLTqpJY2TWVNDOCSf'
access_token_secret = 'SCG14Qvwoou00QYSi4IiUnkQV12tq9uze7Zrm9uuEPAL4'

# Authenticate with the API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

# Open the CSV file containing the tweets
with open('aapdup.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)  # skip the header row
    
    # Read the ID of the last processed tweet from the file, if it exists
    last_processed_id = None
    if os.path.exists('last_processed_tweet_id3.txt'):
        with open('last_processed_tweet_id3.txt', 'r') as id_file:
            last_processed_id = id_file.read().strip()
    else:
        with open('last_processed_tweet_id3.txt', 'w') as f:
            pass 
    
    batch_size = 500  # number of tweets to process in each batch
    batches = [row for row in reader]  # split the data into batches
    batch_count = 0
    processed_count = 0
    processed_tweets = []
    for i in range(0, len(batches), batch_size):
        batch = batches[i:i+batch_size]  # get a batch of tweets
        
        # Process the tweets in the batch
        for row in batch:
            # Extract the tweet ID from the CSV file
            tweet_id = row[0]
            
            # Check if we've already processed this tweet or if we need to resume processing
            if last_processed_id is not None and tweet_id != last_processed_id:
                continue
            else:
                last_processed_id = None  # reset the ID so we process all remaining tweets
            
            # Get the tweet object from Twitter's API
            
            
             
            try:
                tweet = api.get_status(tweet_id, tweet_mode='extended')
                 
                #Extract the retweet count and favorite count from the tweet object      
                retweets = tweet.retweet_count
                favorites = tweet.favorite_count
            
                # Search for replies to the tweet and count the number of results
                query = f"to:{tweet.user.screen_name} since_id:{tweet.id} -filter:retweets"
                replies_count = 0
                for _ in tweepy.Cursor(api.search_tweets, q=query, tweet_mode='extended').items():
                    replies_count += 1
                
                # Append the data to a list of processed tweets
                processed_tweets.append(row + [retweets, favorites, replies_count])
                processed_count += 1
             
                # Save the processed data after every 100 tweets
                if processed_count % 100 == 0:
                    with open('tweets_with_metrics3.csv', 'a', encoding='utf-8', newline='') as g:
                        writer = csv.writer(g)
                        writer.writerows(processed_tweets)
                    processed_tweets = []
            
               # Print progress every 10 batches
                if processed_count % (batch_size * 1) == 0:
                    print(f"{processed_count} tweets processed")
               
            
                    
             
            except tweepy.errors.TweepyException as e:
                if isinstance(e.response,( ConnectionError,ProtocolError)):
                    print("Connection or Protocol Error. Waiting for 5 minutes...")
                    time.sleep(5 * 60)  # Wait for 5 minutes before retrying
                    continue
                if e.response is not None:
                    if e.response.status_code == 88:
                        print("Rate limit exceeded. Waiting for 15 minutes...")
                        time.sleep(15 * 60)  # wait for 15 minutes before retrying
                        continue
                    elif e.response.status_code == 179:
                        print(f"Tweet {tweet_id} is forbidden or unauthorized. Skipping...")
                        continue
                
                with open('last_processed_tweet_id3.txt', 'w') as id_file:
                    id_file.write(tweet_id)
                print(f"Error fetching tweet {tweet_id}: {e}")
                continue
            
                
        # Save the processed data after every batch
        with open('tweets_with_metrics3.csv', 'a', encoding='utf-8', newline='') as g:
            writer = csv.writer(g)
            writer.writerows(processed_tweets)
        processed_tweets = []
        batch_count += 1
        print(f"Batch {batch_count} complete")
```
